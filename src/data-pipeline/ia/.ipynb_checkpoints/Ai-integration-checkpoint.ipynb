```python
import requests
import json
from datetime import datetime
import time

class AIIntegrationTester:
    """
    Suite de tests automatisÃ©s pour valider l'intÃ©gration complÃ¨te du systÃ¨me IA
    Teste tous les endpoints, la qualitÃ© des rÃ©ponses et les performances
    """
    
    def __init__(self, api_base_url='http://localhost:8080/api'):
        self.api_base_url = api_base_url
        self.token = None
        self.test_results = []
    
    def run_all_tests(self):
        """ExÃ©cute tous les tests"""
        print("=" * 80)
        print("SUITE DE TESTS - INTÃ‰GRATION IA")
        print("=" * 80)
        
        # 1. Tests de connectivitÃ©
        print("\nðŸ”Œ TESTS DE CONNECTIVITÃ‰")
        print("-" * 80)
        self.test_backend_health()
        self.test_database_connection()
        
        # 2. Tests d'authentification
        print("\nðŸ” TESTS D'AUTHENTIFICATION")
        print("-" * 80)
        self.test_login()
        
        # 3. Tests des endpoints de recommandations
        print("\nðŸ¤– TESTS DES RECOMMANDATIONS IA")
        print("-" * 80)
        self.test_personalized_recommendations()
        self.test_similar_recipes()
        self.test_trends_detection()
        self.test_for_you_endpoint()
        self.test_explore_endpoint()
        
        # 4. Tests de qualitÃ©
        print("\nâœ… TESTS DE QUALITÃ‰")
        print("-" * 80)
        self.test_recommendation_diversity()
        self.test_response_time()
        self.test_error_handling()
        
        # 5. GÃ©nÃ©rer le rapport
        self.generate_test_report()
    
    def test_backend_health(self):
        """Test la santÃ© du backend"""
        test_name = "Backend Health Check"
        
        try:
            response = requests.get(
                f"{self.api_base_url.replace('/api', '')}/actuator/health",
                timeout=5
            )
            
            if response.status_code == 200:
                data = response.json()
                if data.get('status') == 'UP':
                    self._log_success(test_name, "Backend opÃ©rationnel")
                else:
                    self._log_failure(test_name, f"Status: {data.get('status')}")
            else:
                self._log_failure(test_name, f"Code: {response.status_code}")
                
        except Exception as e:
            self._log_failure(test_name, str(e))
    
    def test_database_connection(self):
        """Test la connexion Ã  la base de donnÃ©es"""
        test_name = "Database Connection"
        
        try:
            # Tenter de rÃ©cupÃ©rer des recettes
            response = requests.get(
                f"{self.api_base_url}/v1/recettes/all",
                timeout=10
            )
            
            if response.status_code == 200:
                recipes = response.json()
                if len(recipes) > 0:
                    self._log_success(test_name, f"{len(recipes)} recettes trouvÃ©es")
                else:
                    self._log_warning(test_name, "Aucune recette dans la base")
            else:
                self._log_failure(test_name, f"Code: {response.status_code}")
                
        except Exception as e:
            self._log_failure(test_name, str(e))
    
    def test_login(self):
        """Test l'authentification"""
        test_name = "Login/Authentication"
        
        try:
            # Utiliser des credentials de test
            response = requests.post(
                f"{self.api_base_url}/v1/auth/login",
                json={
                    "email": "admin@example.com",
                    "motDePasse": "admin123"
                },
                timeout=5
            )
            
            if response.status_code == 200:
                data = response.json()
                if 'token' in data:
                    self.token = data['token']
                    self._log_success(test_name, "Authentification rÃ©ussie")
                else:
                    self._log_failure(test_name, "Token non reÃ§u")
            else:
                self._log_warning(test_name, f"Code: {response.status_code} - CrÃ©er un compte admin")
                
        except Exception as e:
            self._log_failure(test_name, str(e))
    
    def test_personalized_recommendations(self):
        """Test les recommandations personnalisÃ©es"""
        test_name = "Personalized Recommendations"
        
        try:
            user_id = 1  # User de test
            
            start_time = time.time()
            response = requests.get(
                f"{self.api_base_url}/v1/recommendations/personalized/{user_id}",
                timeout=30
            )
            response_time = (time.time() - start_time) * 1000  # en ms
            
            if response.status_code == 200:
                data = response.json()
                recommendations = data.get('recommendations', [])
                
                if len(recommendations) > 0:
                    self._log_success(
                        test_name, 
                        f"{len(recommendations)} recommandations en {response_time:.0f}ms"
                    )
                    
                    # VÃ©rifier la structure des donnÃ©es
                    first_rec = recommendations[0]
                    required_fields = ['id', 'titre', 'typeRecette', 'difficulte']
                    missing_fields = [f for f in required_fields if f not in first_rec]
                    
                    if missing_fields:
                        self._log_warning(
                            test_name, 
                            f"Champs manquants: {missing_fields}"
                        )
                else:
                    self._log_warning(test_name, "Aucune recommandation retournÃ©e")
            else:
                self._log_failure(test_name, f"Code: {response.status_code}")
                
        except Exception as e:
            self._log_failure(test_name, str(e))
    
    def test_similar_recipes(self):
        """Test la recherche de recettes similaires"""
        test_name = "Similar Recipes"
        
        try:
            # RÃ©cupÃ©rer une recette de test
            response = requests.get(
                f"{self.api_base_url}/v1/recettes/all",
                params={'limit': 1},
                timeout=10
            )
            
            if response.status_code == 200:
                recipes = response.json()
                if len(recipes) > 0:
                    recipe_id = recipes[0]['id']
                    
                    # Chercher des recettes similaires
                    start_time = time.time()
                    similar_response = requests.get(
                        f"{self.api_base_url}/v1/recommendations/similar/{recipe_id}",
                        timeout=30
                    )
                    response_time = (time.time() - start_time) * 1000
                    
                    if similar_response.status_code == 200:
                        data = similar_response.json()
                        similar = data.get('similar_recipes', [])
                        
                        if len(similar) > 0:
                            self._log_success(
                                test_name,
                                f"{len(similar)} recettes similaires en {response_time:.0f}ms"
                            )
                        else:
                            self._log_warning(test_name, "Aucune recette similaire trouvÃ©e")
                    else:
                        self._log_failure(test_name, f"Code: {similar_response.status_code}")
                else:
                    self._log_warning(test_name, "Aucune recette pour tester")
            else:
                self._log_failure(test_name, "Impossible de rÃ©cupÃ©rer les recettes")
                
        except Exception as e:
            self._log_failure(test_name, str(e))
    
    def test_trends_detection(self):
        """Test la dÃ©tection de tendances"""
        test_name = "Trends Detection"
        
        try:
            start_time = time.time()
            response = requests.get(
                f"{self.api_base_url}/v1/recommendations/trends",
                timeout=30
            )
            response_time = (time.time() - start_time) * 1000
            
            if response.status_code == 200:
                data = response.json()
                
                if 'ai_analysis' in data:
                    analysis_length = len(data['ai_analysis'])
                    self._log_success(
                        test_name,
                        f"Analyse gÃ©nÃ©rÃ©e ({analysis_length} caractÃ¨res) en {response_time:.0f}ms"
                    )
                else:
                    self._log_warning(test_name, "Analyse IA manquante")
            else:
                self._log_failure(test_name, f"Code: {response.status_code}")
                
        except Exception as e:
            self._log_failure(test_name, str(e))
    
    def test_for_you_endpoint(self):
        """Test l'endpoint 'Pour vous'"""
        test_name = "For You Recommendations"
        
        try:
            user_id = 1
            response = requests.get(
                f"{self.api_base_url}/v1/recommendations/for-you/{user_id}",
                timeout=30
            )
            
            if response.status_code == 200:
                data = response.json()
                
                sections = ['personalized', 'popular', 'newest']
                found_sections = [s for s in sections if s in data]
                
                if len(found_sections) >= 2:
                    self._log_success(
                        test_name,
                        f"{len(found_sections)} sections disponibles"
                    )
                else:
                    self._log_warning(test_name, f"Seulement {len(found_sections)} sections")
            else:
                self._log_failure(test_name, f"Code: {response.status_code}")
                
        except Exception as e:
            self._log_failure(test_name, str(e))
    
    def test_explore_endpoint(self):
        """Test l'endpoint 'DÃ©couverte'"""
        test_name = "Explore Recommendations"
        
        try:
            user_id = 1
            response = requests.get(
                f"{self.api_base_url}/v1/recommendations/explore/{user_id}",
                params={'limit': 20},
                timeout=30
            )
            
            if response.status_code == 200:
                data = response.json()
                recipes = data.get('recipes', [])
                
                if len(recipes) > 0:
                    self._log_success(test_name, f"{len(recipes)} recettes Ã  dÃ©couvrir")
                else:
                    self._log_warning(test_name, "Aucune recette Ã  dÃ©couvrir")
            else:
                self._log_failure(test_name, f"Code: {response.status_code}")
                
        except Exception as e:
            self._log_failure(test_name, str(e))
    
    def test_recommendation_diversity(self):
        """Test la diversitÃ© des recommandations"""
        test_name = "Recommendation Diversity"
        
        try:
            user_id = 1
            response = requests.get(
                f"{self.api_base_url}/v1/recommendations/personalized/{user_id}",
                timeout=30
            )
            
            if response.status_code == 200:
                data = response.json()
                recommendations = data.get('recommendations', [])
                
                if len(recommendations) > 0:
                    # Analyser la diversitÃ©
                    types = set()
                    cuisines = set()
                    difficulties = set()
                    
                    for rec in recommendations:
                        types.add(rec.get('typeRecette'))
                        if rec.get('cuisine'):
                            cuisines.add(rec.get('cuisine'))
                        difficulties.add(rec.get('difficulte'))
                    
                    diversity_score = (len(types) + len(cuisines) + len(difficulties)) / 9
                    
                    if diversity_score > 0.5:
                        self._log_success(
                            test_name,
                            f"Bonne diversitÃ© (score: {diversity_score:.2f})"
                        )
                    else:
                        self._log_warning(
                            test_name,
                            f"DiversitÃ© limitÃ©e (score: {diversity_score:.2f})"
                        )
                else:
                    self._log_warning(test_name, "Aucune recommandation pour tester")
            else:
                self._log_failure(test_name, f"Code: {response.status_code}")
                
        except Exception as e:
            self._log_failure(test_name, str(e))
    
    def test_response_time(self):
        """Test les temps de rÃ©ponse"""
        test_name = "Response Time Performance"
        
        try:
            endpoints = [
                f"{self.api_base_url}/v1/recommendations/personalized/1",
                f"{self.api_base_url}/v1/recommendations/similar/1",
                f"{self.api_base_url}/v1/recommendations/for-you/1"
            ]
            
            times = []
            
            for endpoint in endpoints:
                start = time.time()
                response = requests.get(endpoint, timeout=30)
                elapsed = (time.time() - start) * 1000
                times.append(elapsed)
            
            avg_time = sum(times) / len(times)
            max_time = max(times)
            
            if avg_time < 2000:  # Moins de 2 secondes
                self._log_success(
                    test_name,
                    f"Temps moyen: {avg_time:.0f}ms (max: {max_time:.0f}ms)"
                )
            elif avg_time < 5000:  # Moins de 5 secondes
                self._log_warning(
                    test_name,
                    f"Temps moyen acceptable: {avg_time:.0f}ms"
                )
            else:
                self._log_failure(
                    test_name,
                    f"Temps trop Ã©levÃ©: {avg_time:.0f}ms"
                )
                
        except Exception as e:
            self._log_failure(test_name, str(e))
    
    def test_error_handling(self):
        """Test la gestion des erreurs"""
        test_name = "Error Handling"
        
        try:
            # Test avec un ID invalide
            response = requests.get(
                f"{self.api_base_url}/v1/recommendations/personalized/99999",
                timeout=10
            )
            
            # On s'attend Ã  une rÃ©ponse mÃªme si pas d'historique
            if response.status_code in [200, 404]:
                self._log_success(test_name, "Gestion des IDs invalides OK")
            else:
                self._log_warning(
                    test_name,
                    f"Code inattendu pour ID invalide: {response.status_code}"
                )
                
        except Exception as e:
            self._log_failure(test_name, str(e))
    
    def _log_success(self, test_name, message):
        """Log un test rÃ©ussi"""
        result = {
            'test': test_name,
            'status': 'SUCCESS',
            'message': message,
            'timestamp': datetime.now().isoformat()
        }
        self.test_results.append(result)
        print(f"   âœ… {test_name}: {message}")
    
    def _log_failure(self, test_name, message):
        """Log un test Ã©chouÃ©"""
        result = {
            'test': test_name,
            'status': 'FAILURE',
            'message': message,
            'timestamp': datetime.now().isoformat()
        }
        self.test_results.append(result)
        print(f"   âŒ {test_name}: {message}")
    
    def _log_warning(self, test_name, message):
        """Log un avertissement"""
        result = {
            'test': test_name,
            'status': 'WARNING',
            'message': message,
            'timestamp': datetime.now().isoformat()
        }
        self.test_results.append(result)
        print(f"   âš ï¸ {test_name}: {message}")
    
    def generate_test_report(self):
        """GÃ©nÃ¨re un rapport de test"""
        print("\n" + "=" * 80)
        print("RAPPORT DE TESTS")
        print("=" * 80)
        
        total = len(self.test_results)
        successes = len([r for r in self.test_results if r['status'] == 'SUCCESS'])
        failures = len([r for r in self.test_results if r['status'] == 'FAILURE'])
        warnings = len([r for r in self.test_results if r['status'] == 'WARNING'])
        
        print(f"\nðŸ“Š RÃ‰SULTATS:")
        print(f"   Total: {total} tests")
        print(f"   âœ… SuccÃ¨s: {successes} ({successes/total*100:.1f}%)")
        print(f"   âŒ Ã‰checs: {failures} ({failures/total*100:.1f}%)")
        print(f"   âš ï¸ Avertissements: {warnings} ({warnings/total*100:.1f}%)")
        
        # Sauvegarder le rapport JSON
        with open('test_report.json', 'w', encoding='utf-8') as f:
            json.dump({
                'summary': {
                    'total': total,
                    'successes': successes,
                    'failures': failures,
                    'warnings': warnings,
                    'success_rate': successes / total if total > 0 else 0
                },
                'tests': self.test_results,
                'generated_at': datetime.now().isoformat()
            }, f, indent=2, ensure_ascii=False)
        
        print(f"\nðŸ’¾ Rapport sauvegardÃ©: test_report.json")
        
        # Ã‰valuation finale
        if failures == 0:
            print("\nðŸŽ‰ TOUS LES TESTS SONT PASSÃ‰S!")
        elif failures < 3:
            print("\nâš ï¸ QUELQUES TESTS ONT Ã‰CHOUÃ‰ - ATTENTION")
        else:
            print("\nðŸš¨ PLUSIEURS TESTS ONT Ã‰CHOUÃ‰ - ACTION REQUISE")


# ExÃ©cution
if __name__ == "__main__":
    print("=" * 80)
    print("DÃ‰MARRAGE DES TESTS AUTOMATISÃ‰S")
    print("=" * 80)
    
    API_URL = input("\nURL de l'API (dÃ©faut: http://localhost:8080/api): ").strip()
    if not API_URL:
        API_URL = "http://localhost:8080/api"
    
    tester = AIIntegrationTester(api_base_url=API_URL)
    tester.run_all_tests()
    
    print("\n" + "=" * 80)
    print("âœ… TESTS TERMINÃ‰S")
    print("=" * 80)
```
